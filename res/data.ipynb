```python
import pandas as pd


def dummy_data(num_points):
    df = pd.read_csv("030522_1542_4_0_testing0.csv", skiprows=5, names = ["time_point", "time_us", "value"])
    buffer = '"trace_begun/r/n"'

    for idx, row in df.iterrows():
        if idx <= num_points:
            buffer += f'{row.loc["time_point"]}, {row.loc["time_us"]}," {row.loc["value"]} /r/n"'
    return buffer

res = dummy_data(10)
print(res)
```

```python

class DummyData:
    def __init__(self):
        self.df = pd.read_csv("030522_1542_4_0_testing0.csv", skiprows=5, names = ["time_point", "time_us", "value"])

    def get_row(self, idx):
        return f'{idx}, {self.df.time_us[idx]}," {self.df.value[idx]} /r/n"'
        

```

```python

data = DummyData()
buffer = ""

for i in range(0, 100):
    buffer += data.get_row(i)

print(buffer)
```
